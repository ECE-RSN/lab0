### VirtualBox instructions for Ubuntu (Intel-based Macs and Windows):  

1. Download the .iso for Ubuntu 20.04 LTS (long term service).  https://releases.ubuntu.com/focal/ 

2. Install Virtual Box or other virtualization software: https://www.virtualbox.org/ 

3. Use these instructions to set up a Ubuntu machine with Virtual Box: https://ubuntu.com/tutorials/how-to-run-ubuntu-desktop-on-a-virtual-machine-using-virtualbox#1-overview 

4. I recommend 2048+ MB RAM, 25+ GB of hard disk space, and it doesn’t matter if you choose fixed or dynamic virtual drive 

5. Be sure to check out the last page of instructions to understand how to resize the window! 

6. After you complete installation, make sure you “eject” the .iso file from the virtual machine using the VM settings, or it will ask if you want to install again. 


### UTM instructions for Ubuntu (Apple M1/M2 ARM processor Macs):  

1. Download the .iso for Ubuntu 20.04 daily build: https://cdimage.ubuntu.com/focal/daily-live/pending/  

2. Install UTM from the Apple store (price was 10USD in Fall 2022) or from getutm.app (free in May 2023) 

3. Use these instructions to set up a Ubuntu machine with Virtual Box: https://ubuntu.com/tutorials/how-to-run-ubuntu-desktop-on-a-virtual-machine-using-virtualbox  

4. I recommend 4096+ MB RAM, 25+ GB of hard disk space, and dynamic virtual drive 

5. After you complete installation, make sure you “eject” the .iso file from the virtual machine using the VM settings, or it will ask if you want to install again. 
     
### Using Windows Terminal (at your own risk!):

Here are some helpful links if you want to try to use Windows Terminal for the course:

Setting up ROS in Windows through WSL2: https://jackkawell.wordpress.com/2020/06/12/ros-wsl2/ 

Installing ROS Noetic on WSL2: https://github.com/ishkapoor2000/Install_ROS_Noetic_On_WSL

Connecting USB Devices to WSL: https://learn.microsoft.com/en-us/windows/wsl/connect-usb

Creating Bash aliases in WSL: https://microcollaborative.atlassian.net/wiki/spaces/DSC/pages/147521540/Creating+aliases+in+Windows+Subsystem+for+Linux

For the most part, WSL should be similar to running Ubuntu natively, but it is important to understand how to navigate the file system via WSL since it's an emulation of ubuntu on top of the windows file system. So the "~" directory is still your home directory but the windows file system is found at "/mnt/c/". Anything students download and put on their desktop or into downloads or documents can be found at "/mnt/c/Users/<username>/" whether it's "Documents", "Downloads", or "Desktop".
