### VirtualBox instructions for Ubuntu (Intel-based Macs and Windows):  

1. Download the .iso for Ubuntu 24.04 Desktop (AMD): https://cdimage.ubuntu.com/noble/daily-live/current/

2. Install Virtual Box or other virtualization software: https://www.virtualbox.org/ Use VirtualBox if you want to use the course install. 

3. Use these instructions to set up a Ubuntu machine with Virtual Box: https://ubuntu.com/tutorials/how-to-run-ubuntu-desktop-on-a-virtual-machine-using-virtualbox#1-overview 

4. I recommend 4096+ MB RAM, 25+ GB of hard disk space, and it doesn’t matter if you choose fixed or dynamic virtual drive. The course VM has 8 GB of RAM and 48 GB of hard drive space.

5. VERY IMPORTANT: When specifying the location of the .iso file, CHECK "Skip Unattended Installation" so your username gets added to the sudoers list (This is critical so you can have administrator privileges on the machine). The course VM has username paws and password imahusky.

6. Be sure to check out the last page of instructions to understand how to resize the window! 

7. After you complete installation, make sure you “eject” the .iso file from the virtual machine using the VM settings, or it will ask if you want to install again. 


### UTM instructions for Ubuntu (Apple M* ARM processor Macs):  

1. Download a 64-bit ARM desktop .iso: https://cdimage.ubuntu.com/noble/daily-live/current/

2. Install virtualization software such as UTM (from the Apple store, price was 10USD in Fall 2022) or from the website getutm.app (free in May 2023) 

3. Use these instructions to set up a Ubuntu machine with Virtual Box: https://ubuntu.com/tutorials/how-to-run-ubuntu-desktop-on-a-virtual-machine-using-virtualbox  

4. I recommend 4096+ MB RAM, 25+ GB of hard disk space, and dynamic virtual drive 

5. After you complete installation, make sure you “eject” the .iso file from the virtual machine using the VM settings, or it will ask if you want to install again.
